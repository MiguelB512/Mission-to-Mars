```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# Set up Splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

# With these two lines of code, we are creating an instance of a Splinter browser. This means that we're prepping our 
# automated browser. We're also specifying that we'll be using Chrome as our browser. **executable_path is unpacking the 
# dictionary we've stored the path in – think of it as unpacking a suitcase. headless=False means that all of the browsers 
# actions will be displayed in a Chrome window so we can see them.
```

```python
# Visit the Quotes to Scrape site. Tell splinter which website we want to visit. 
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
# Parse the HTML using beautiful soup (soup). We are using an html parser, but other parsers can be used. 
html = browser.html
html_soup = soup(html, 'html.parser')
```

```python
# Scrape the Title from the webpage.
title = html_soup.find('h2').text
title
```

```python
# Scrape the top ten tags
tag_box = html_soup.find('div', class_='tags-box')

# tag_box
tags = tag_box.find_all('a', class_='tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
# Make a look to go through pages 1-6 (5 pages)
for x in range(1, 6):
   
   # Connect the browser and tell it that the file is HTML 
   html = browser.html
    
   # Parse the html using beautifulsoup  
   quote_soup = soup(html, 'html.parser')
   
   # Scrape for 'span' that have the class 'text' and add them to 'quotes'
   quotes = quote_soup.find_all('span', class_='text')
   
   # Begin looping throug the quotes, print them out along with the page and '---' 
   for quote in quotes:
      print('page:', x, '----------')
      print(quote.text)
   browser.links.find_by_partial_text('Next').click()
```

```python
# My attempt at extracting the links from the books.toscrape website
url = 'http://books.toscrape.com/'
browser.visit(url)
html = browser.html
html_soup = soup(html, 'html.parser')
book_list = html_soup.find_all('ul', class_= 'nav-list')

for book in book_list:
    
    
    try:
        text_list = book.find_all('li')
        
        link = book.a['href']
        
        if (text_list and link):
            print('-------------')
            print(text_list)
            print(link)
    
    except AttributeError as e:
        print(e)
```
